```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder
from matplotlib import pyplot as plt

import torch
import torch.nn as nn
from pytorch_tabnet.tab_model import TabNetClassifier
```

 TabNet은 입력으로 Categorical변수를 Embedding 하기 때문에, Categorical 변수라는 것을 지정해주어야 합니다. 따라서, cat_idxs와 cat_dims를 저장해둡니다.

```python
train_df.nunique()
```

```python
train_df.columns
```

```python
target = 'depvar' #train_df의 마지막 칼럼이 target

nunique = train_df.nunique()
types = train_df.dtypes

categorical_columns = ['home_ownership', 'verification_status', 'purpose', 'initial_list_status']
categorical_dims =  {'home_ownership': 6, 'verification_status': 3, 'purpose' : 13, 'initial_list_status': 2}


# Categorical Embedding을 위해 Categorical 변수의 차원과 idxs를 담음.
unused_feat = ['Set']
features = [ col for col in train_df.columns if col not in unused_feat+[target]] 
cat_idxs = [ i for i, f in enumerate(features) if f in categorical_columns]
cat_dims = [ categorical_dims[f] for i, f in enumerate(features) if f in categorical_columns]



# X_train = train[features].values[train_indices]
# y_train = train[target].values[train_indices]

# X_valid = train[features].values[valid_indices]
# y_valid = train[target].values[valid_indices]

# X_test = train[features].values[test_indices]
# y_test = train[target].values[test_indices]
```

```python
cat_idxs
```

```python
cat_dims
```

```python

DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

clf = TabNetClassifier(cat_idxs=cat_idxs,
                       cat_dims=cat_dims,
                       cat_emb_dim=10,
                       optimizer_fn=torch.optim.Adam,
                       optimizer_params=dict(lr=1e-2),
                       scheduler_params={"step_size":50,
                                         "gamma":0.9},
                       scheduler_fn=torch.optim.lr_scheduler.StepLR,
                       mask_type='sparsemax', # "sparsemax", entmax
                       device_name = DEVICE
                      )
```

```python
max_epochs = 15

clf.fit(
    X_train=X_train, y_train=y_train,
    eval_set=[(X_train, y_train), (X_valid, y_valid)],
    eval_name=['train', 'valid'],
    eval_metric=['f1_macro'],
    max_epochs=max_epochs , patience=20,
    batch_size=1024, virtual_batch_size=128,
    num_workers=0,
    weights=1,
    drop_last=False,
)
```
